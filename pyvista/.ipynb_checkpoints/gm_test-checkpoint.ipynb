Just testing new GM kinetics.

```python
import numpy as np
import scipy.sparse as sp
import matplotlib.pyplot as plt
```

```python
from forward import step_se_gm

nx = 25
du = 9.45e-4
dv = .27
r = .001
a = 200
mu = 2.5
dt = 1e-8
dx = 0.05

# initialize near steady-state solution
u = np.ones(nx)*(a)
u[4] *= 1.1 # perturbation to test
v = np.ones(nx)*(a)

# finite difference for laplacian
# boundaries currently periodic 
lap = sp.diags([-2,1,1,1,1], offsets=[0,1,-1,nx-1,1-nx], shape=(nx,nx))

# identity matrix
iden = sp.eye(nx)


# integrate over time, solving using GMRES at each timestep
niter = 1000

u_stored = np.zeros((nx, niter+1))
u_stored[:,0] = u

v_stored = np.zeros((nx, niter+1))
v_stored[:,0] = v

for i in range(niter):
    # Run GMRES to solve for next timestep
    u, v = step_se_gm(u, v, r, mu, a, du, dv, lap, dx, nx, dt)

    # store for later animation
    u_stored[:,i+1] = u
    v_stored[:,i+1] = v
```

```python
plt.plot(u_stored[:,50])
plt.show()

# from animation import animate_2d_array

# animate_2d_array(u_stored[:,::10], interval=1)

# plt.plot(u_stored[:,-1])
```
